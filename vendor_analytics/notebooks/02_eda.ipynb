```python
import pandas as pd
from sqlalchemy import create_engine

engine = create_engine('mysql+pymysql://root:mysql123@localhost/vendor_analytics')

tables = ['raw_purchases', 'raw_end_inventory', 
          'raw_purchase_prices', 'raw_vendor_invoice']

for table in tables:
    df = pd.read_sql(f"SELECT * FROM {table} LIMIT 3", engine)
    print(f"\n{'='*60}")
    print(f"TABLE: {table}")
    print(f"Columns: {list(df.columns)}")
    print(df.head(3))
```

```python
for table in tables:
    df = pd.read_sql(f"SELECT * FROM {table}", engine)
    print(f"\n{'='*60}")
    print(f"TABLE: {table}  |  Rows: {len(df):,}")
    print("\nNull counts:")
    print(df.isnull().sum()[df.isnull().sum() > 0])
    print("\nData types:")
    print(df.dtypes)
```

```python
df = pd.read_sql("SELECT DISTINCT VendorName FROM raw_purchases ORDER BY VendorName", engine)
print(f"Unique vendor names: {len(df)}")
print(df.head(30))
```
