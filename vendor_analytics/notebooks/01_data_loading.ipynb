```python
import pandas as pd
from sqlalchemy import create_engine

engine = create_engine('mysql+pymysql://root:mysql123@localhost/vendor_analytics')

# Load beginning inventory
print("Loading beginning inventory...")
df_begin = pd.read_csv('../data/begin_inventory.csv')
print(f"Rows in file: {len(df_begin)}")
print(df_begin.head())  # shows first 5 rows so you can see what it looks like
```

```python
df_begin.to_sql('raw_inventory_begin', engine, if_exists='replace', index=False)
print("✅ Beginning inventory loaded into MySQL!")
```

```python
import os

# -----------------------------------------------
# CHANGE THESE FILE NAMES to match yours exactly
# -----------------------------------------------
files = {
    'raw_end_inventory'  : '../data/end_inventory.csv',
    'raw_purchase_prices': '../data/purchase_prices.csv',
    'raw_vendor_invoice' : '../data/vendor_invoice.csv',
    'raw_purchases'      : '../data/purchases.csv',
}

# Load each file
for table_name, file_path in files.items():
    print(f"Loading {file_path} ...")
    
    df = pd.read_csv(file_path)
    df.to_sql(table_name, engine, if_exists='replace', index=False)
    
    print(f"  ✅ {table_name} → {len(df):,} rows loaded")

print("\n🎉 All files loaded!")
```

```python
print("Loading sales file (will take 3-5 minutes)...")

chunk_num = 0
for chunk in pd.read_csv('../data/sales.csv', chunksize=10000):
    chunk.to_sql('raw_sales', engine, if_exists='append', index=False)
    chunk_num += 1
    if chunk_num % 20 == 0:
        print(f"  ...{chunk_num * 10000:,} rows loaded so far")

print("✅ Sales file done!")
```
