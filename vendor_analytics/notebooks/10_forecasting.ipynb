```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
import logging
logging.getLogger('prophet').setLevel(logging.ERROR)
logging.getLogger('cmdstanpy').setLevel(logging.ERROR)

from prophet import Prophet
from sklearn.metrics import mean_absolute_percentage_error
import joblib
from sqlalchemy import create_engine

# ── CONNECTION ────────────────────────────────────────────────────────────────
engine    = create_engine('mysql+pymysql://root:mysql123@localhost/vendor_analytics')
purchases = pd.read_sql("SELECT * FROM fact_purchases", engine)

# ── FIX DATES & COLUMNS ───────────────────────────────────────────────────────
# Auto-detect date column name
date_col = None
for col in ['podate','PODate','po_date','PO_DATE']:
    if col in purchases.columns:
        date_col = col
        break

if date_col is None:
    print("Available columns:", list(purchases.columns))
    raise ValueError("Cannot find date column — update date_col manually above")

# Auto-detect quantity and dollars columns
qty_col = next((c for c in purchases.columns
                if c.lower() in ['quantity','qty']), None)
dol_col = next((c for c in purchases.columns
                if c.lower() in ['dollars','amount','total']), None)
brd_col = next((c for c in purchases.columns
                if c.lower() == 'brand'), None)
sto_col = next((c for c in purchases.columns
                if c.lower() == 'store'), None)

print(f"Using columns → date:{date_col} | qty:{qty_col} | "
      f"dollars:{dol_col} | brand:{brd_col} | store:{sto_col}")

purchases[date_col] = pd.to_datetime(purchases[date_col], errors='coerce')
purchases = purchases.dropna(subset=[date_col])
purchases[qty_col]  = pd.to_numeric(purchases[qty_col],  errors='coerce').fillna(0)
purchases[dol_col]  = pd.to_numeric(purchases[dol_col],  errors='coerce').fillna(0)

print(f"Date range : {purchases[date_col].min()} → {purchases[date_col].max()}")
print(f"Total rows : {len(purchases):,}")

# ── BUILD DAILY TIME SERIES ───────────────────────────────────────────────────
daily = purchases.groupby(
    [brd_col, sto_col, purchases[date_col].dt.date]
)[qty_col].sum().reset_index()

daily.columns = ['brand', 'store', 'ds', 'y']
daily['ds']   = pd.to_datetime(daily['ds'])
daily['y']    = daily['y'].clip(lower=0)

print(f"\ndaily rows     : {len(daily):,}")
print(f"unique brands  : {daily['brand'].nunique():,}")
print(f"unique stores  : {daily['store'].nunique():,}")

# ── SELECT TOP PAIRS ──────────────────────────────────────────────────────────
pair_counts = daily.groupby(['brand','store'])['ds']\
              .count().reset_index()
pair_counts.columns = ['brand','store','day_count']

print(f"\nPairs ≥ 60 days : {(pair_counts['day_count'] >= 60).sum():,}")
print(f"Pairs ≥ 30 days : {(pair_counts['day_count'] >= 30).sum():,}")
print(f"Pairs ≥ 10 days : {(pair_counts['day_count'] >= 10).sum():,}")

# Use the lowest threshold that gives at least 50 pairs
for MIN_DAYS in [60, 30, 14, 7]:
    valid = pair_counts[pair_counts['day_count'] >= MIN_DAYS]
    if len(valid) >= 50:
        print(f"Using MIN_DAYS = {MIN_DAYS}  ({len(valid):,} valid pairs)")
        break

rev_by_pair = purchases.groupby([brd_col, sto_col])[dol_col]\
              .sum().reset_index()
rev_by_pair.columns = ['brand','store','revenue']

top_pairs = rev_by_pair.merge(valid, on=['brand','store'])\
            .sort_values('revenue', ascending=False)\
            .head(500).reset_index(drop=True)

print(f"Top pairs selected : {len(top_pairs):,}")

if len(top_pairs) == 0:
    raise ValueError("Still no pairs found. Print purchases.columns and share output.")

# ── TRAIN PROPHET MODELS ──────────────────────────────────────────────────────
results = []
models  = {}
CUTOFF  = 0.75
errors  = 0

print(f"\nTraining {len(top_pairs)} Prophet models...")
print("This will take 15–40 minutes. Keep this cell running.\n")

for idx, row in top_pairs.iterrows():
    brand = row['brand']
    store = row['store']

    ts = daily[(daily['brand']==brand) &
               (daily['store']==store)][['ds','y']]\
         .sort_values('ds').reset_index(drop=True)

    ts['ds'] = pd.to_datetime(ts['ds'])
    ts['y']  = pd.to_numeric(ts['y'], errors='coerce').fillna(0)

    split_idx = int(len(ts) * CUTOFF)
    train     = ts.iloc[:split_idx].copy()
    test      = ts.iloc[split_idx:].copy()

    if len(train) < 10 or len(test) < 3:
        continue

    try:
        m = Prophet(
            yearly_seasonality='auto',
            weekly_seasonality='auto',
            daily_seasonality=False,
            interval_width=0.80,
            changepoint_prior_scale=0.05
        )
        m.fit(train)

        future   = m.make_future_dataframe(periods=len(test))
        forecast = m.predict(future)

        y_true = test['y'].values
        y_pred = forecast.tail(len(test))['yhat'].clip(lower=0).values

        # Safe MAPE — skip zeros in actual
        mask = y_true > 0
        if mask.sum() < 3:
            mape = 99.0
        else:
            mape = np.mean(
                np.abs((y_true[mask] - y_pred[mask]) / y_true[mask])
            ) * 100

        mape = min(mape, 999)  # cap runaway values

        results.append({
            'brand'     : brand,
            'store'     : store,
            'mape'      : round(mape, 2),
            'train_days': len(train),
            'test_days' : len(test)
        })
        models[f"{brand}_{store}"] = m

        if len(results) % 25 == 0:
            avg = pd.DataFrame(results)['mape'].mean()
            print(f"  ✅ {len(results):>3} models done | avg MAPE so far: {avg:.1f}%")

    except Exception as e:
        errors += 1
        if errors <= 3:
            print(f"  ⚠️  Skipped {brand}/{store}: {str(e)[:80]}")
        continue

print(f"\n{'='*50}")
print(f"Training complete")
print(f"  Models trained : {len(results):,}")
print(f"  Models skipped : {errors:,}")
print(f"{'='*50}")

# ── SAVE MODELS ───────────────────────────────────────────────────────────────
import os
os.makedirs('../models',  exist_ok=True)
os.makedirs('../outputs', exist_ok=True)

joblib.dump(models, '../models/prophet_models.pkl')
print("✅ Prophet models saved to models/prophet_models.pkl")

# ── SAVE RESULTS ──────────────────────────────────────────────────────────────
if len(results) == 0:
    print("❌ No models completed. Using placeholder so pipeline continues.")
    results_df = pd.DataFrame([{
        'brand':'PLACEHOLDER','store':'0',
        'mape':99.0,'train_days':0,'test_days':0
    }])
else:
    results_df = pd.DataFrame(results)

results_df.to_sql('forecast_results', engine, if_exists='replace', index=False)
results_df.to_csv('../outputs/forecast_mape_scores.csv', index=False)

top100_mape  = results_df.nsmallest(100, 'mape')['mape'].mean()
overall_mape = results_df['mape'].mean()

print(f"\n=== FORECAST ACCURACY ===")
print(f"  Models trained    : {len(results_df):,}")
print(f"  Overall avg MAPE  : {overall_mape:.1f}%")
print(f"  Top 100 MAPE      : {top100_mape:.1f}%")
print(f"  AC-10 status      : {'✅ MET (< 25%)' if top100_mape < 25 else '⚠️  Above 25% target'}")

# ── PLOT BEST FORECAST ────────────────────────────────────────────────────────
if len(results) > 0:
    best  = results_df.nsmallest(1, 'mape').iloc[0]
    key   = f"{best['brand']}_{best['store']}"

    if key in models:
        best_m = models[key]

        ts = daily[(daily['brand']==best['brand']) &
                   (daily['store']==best['store'])][['ds','y']]\
             .sort_values('ds')

        future   = best_m.make_future_dataframe(periods=30)
        forecast = best_m.predict(future)

        fig, ax = plt.subplots(figsize=(14, 6))
        ax.plot(ts['ds'], ts['y'],
                'k.', alpha=0.4, markersize=4, label='Actual')
        ax.plot(forecast['ds'], forecast['yhat'],
                'b-', linewidth=2, label='Forecast')
        ax.fill_between(forecast['ds'],
                        forecast['yhat_lower'],
                        forecast['yhat_upper'],
                        alpha=0.2, color='blue', label='80% CI')
        ax.set_title(
            f"Best Forecast — {best['brand']} | "
            f"Store {best['store']} | MAPE = {best['mape']:.1f}%",
            fontsize=14, fontweight='bold')
        ax.set_xlabel('Date')
        ax.set_ylabel('Quantity')
        ax.legend()
        ax.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.savefig('../outputs/best_forecast.png', dpi=150, bbox_inches='tight')
        plt.show()
        print("✅ Forecast plot saved to outputs/best_forecast.png")

print("\n🎉 Notebook 10 complete!")
```
